````python
import pandas as pd

# Define a list named csv_files containing the names of all CSV files to be inspected
csv_files = [
    'website_sessions.csv',
    'website_pageviews.csv',
    'orders.csv',
    'order_items.csv',
    'order_item_refunds.csv',
    'products.csv',
    'maven_fuzzy_factory_data_dictionary.csv'
]

# Initialize an empty dictionary named dataframes to store the loaded DataFrames
dataframes = {}

# Loop through each file in the csv_files list
for file_name in csv_files:
    # Print a header indicating the current file being inspected
    print(f"---\nInspecting {file_name}\n---")

    file_path = f'/content/{file_name}'

    try:
        # Use pd.read_csv() to load the current CSV file into a DataFrame
        df = pd.read_csv(file_path)

        # Store the loaded DataFrame in the dataframes dictionary
        # using the file name (without the '.csv' extension) as the key
        df_name = file_name.replace('.csv', '')
        dataframes[df_name] = df

        # Print the first few rows of the DataFrame
        print("\nHead of DataFrame:")
        print(df.head())

        # Print a summary of the DataFrame including column data types and non-null values
        print("\nInfo of DataFrame:")
        df.info()

        # Print descriptive statistics for all columns (including non-numeric)
        print("\nDescriptive Statistics:")
        print(df.describe(include='all'))

        # If the current file is 'maven_fuzzy_factory_data_dictionary.csv',
        # temporarily set pandas display options to show all rows and columns,
        # then print the entire DataFrame, and finally reset the display options to default.
        if file_name == 'maven_fuzzy_factory_data_dictionary.csv':
            print("\nFull Data Dictionary:")
            with pd.option_context('display.max_rows', None, 'display.max_columns', None):
                print(df)

    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
    except Exception as e:
        print(f"An error occurred while processing {file_name}: {e}")

    # Print a separator line for better readability between file inspections
    print("\n" + "="*50 + "\n")

print("All specified CSV files have been loaded and inspected.")
````

## Formulate Statistical Hypotheses

### Subtask:
Translate the chosen business hypothesis into a Null Hypothesis (H₀) and an Alternative Hypothesis (H₁) for statistical testing. Specify a significance level (α), as recommended in the brief (α = 0.05).


## Formulate Statistical Hypotheses

### Business Hypothesis Chosen:
From the hackathon brief, a suggested business hypothesis is: "Paid marketing channels increase conversion rate."

### Statistical Hypotheses:

1.  **Null Hypothesis (H₀):** There is no significant difference in conversion rates between paid marketing channels and other marketing channels. (i.e., Paid marketing channels do not increase conversion rate).
    
    *Mathematically:* $\mu_{\text{paid}} = \mu_{\text{other}}$ or $\mu_{\text{paid}} \le \mu_{\text{other}}$

2.  **Alternative Hypothesis (H₁):** Paid marketing channels lead to a significantly higher conversion rate compared to other marketing channels.
    
    *Mathematically:* $\mu_{\text{paid}} > \mu_{\text{other}}$

### Significance Level:

*   **Alpha (α):** 0.05 (as recommended in the brief).

This means we are willing to accept a 5% chance of incorrectly rejecting the Null Hypothesis (Type I error).

## Data Preprocessing and Exploration

### Subtask:
Perform data quality checks, calculate descriptive statistics, and begin initial data exploration for the 'website_sessions' and 'orders' datasets, focusing on elements relevant to marketing channels and conversion rates.


**Reasoning**:
First, I will access the `website_sessions` and `orders` DataFrames from the `dataframes` dictionary and convert their `created_at` columns to datetime objects. This is crucial for accurate time-based analysis and merging operations.



````python
website_sessions_df = dataframes['website_sessions'].copy()
orders_df = dataframes['orders'].copy()

# Convert 'created_at' columns to datetime objects
website_sessions_df['created_at'] = pd.to_datetime(website_sessions_df['created_at'])
orders_df['created_at'] = pd.to_datetime(orders_df['created_at'])

print("Converted 'created_at' columns to datetime for website_sessions_df and orders_df.")
print(f"website_sessions_df['created_at'] dtype: {website_sessions_df['created_at'].dtype}")
print(f"orders_df['created_at'] dtype: {orders_df['created_at'].dtype}")
````

**Reasoning**:
The next step is to merge the `website_sessions_df` and `orders_df` to combine session and order information. A left merge will ensure all sessions are retained, and order details are included only where a conversion occurred.



````python
df_merged = pd.merge(
    website_sessions_df,
    orders_df[['website_session_id', 'order_id']],
    on='website_session_id',
    how='left'
)

print("Merged website_sessions_df with orders_df to create df_merged.")
print("Head of df_merged:")
print(df_merged.head())
````

**Reasoning**:
Now that the `website_sessions_df` and `orders_df` are merged, I will create a new binary column named 'converted' in `df_merged`. This column will indicate whether a session resulted in an order (1) or not (0) based on the presence of an `order_id`.



````python
df_merged['converted'] = df_merged['order_id'].notna().astype(int)

print("Added 'converted' column to df_merged.")
print("Head of df_merged with 'converted' column:")
print(df_merged.head())
````

**Reasoning**:
I will categorize marketing channels by creating a new column `marketing_channel_type` in `df_merged`. Sessions with `utm_source` as 'gsearch' or 'bsearch' will be labeled as 'paid', as well as sessions where `utm_source` is missing but `http_referer` indicates 'gsearch.com'. All other sessions will be categorized as 'other' to fulfill the subtask's requirements for distinguishing marketing channels.



````python
df_merged['marketing_channel_type'] = 'other'

# Categorize 'gsearch' and 'bsearch' utm_source as 'paid'
df_merged.loc[df_merged['utm_source'].isin(['gsearch', 'bsearch']), 'marketing_channel_type'] = 'paid'

# Handle NaN utm_source based on http_referer for 'paid'
df_merged.loc[
    (df_merged['utm_source'].isna()) &
    (df_merged['http_referer'].str.contains('gsearch.com', na=False)),
    'marketing_channel_type'
] = 'paid'

# Inspect the unique values and counts of the 'utm_source' column
print("Unique values and counts for 'utm_source' before categorization:")
print(website_sessions_df['utm_source'].value_counts(dropna=False))

print("\nMarketing channel type distribution:")
print(df_merged['marketing_channel_type'].value_counts())

print("\nHead of df_merged with 'marketing_channel_type' column:")
print(df_merged.head())
````

**Reasoning**:
Now that the marketing channels are categorized and conversion status is available, I will calculate descriptive statistics for the 'converted' column, grouped by 'marketing_channel_type'. This will provide an initial understanding of conversion rates for each channel, fulfilling part of the subtask requirements.



````python
conversion_rates = df_merged.groupby('marketing_channel_type')['converted'].agg(['count', 'sum', 'mean']).reset_index()
conversion_rates.rename(columns={'count': 'total_sessions', 'sum': 'total_conversions', 'mean': 'conversion_rate'}, inplace=True)

print("Conversion rates by marketing channel type:")
print(conversion_rates)
````

**Reasoning**:
To visualize the distribution of marketing channel types, I will create a bar plot showing the count of sessions for each category ('paid' vs. 'other'). This addresses instruction 8 of the subtask.



````python
import matplotlib.pyplot as plt

# Visualize the distribution of marketing_channel_type
plt.figure(figsize=(8, 6))
df_merged['marketing_channel_type'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Distribution of Marketing Channel Types')
plt.xlabel('Marketing Channel Type')
plt.ylabel('Number of Sessions')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("Bar plot showing the distribution of marketing channel types has been generated.")
````

**Reasoning**:
To complete the subtask, I will now visualize the conversion rates for 'paid' vs. 'other' marketing channels using a bar plot, based on the `conversion_rates` DataFrame, which will highlight the difference in conversion success.



````python
import matplotlib.pyplot as plt

# Visualize the conversion rates for 'paid' vs. 'other' marketing channels
plt.figure(figsize=(8, 6))
plt.bar(conversion_rates['marketing_channel_type'], conversion_rates['conversion_rate'], color=['skyblue', 'lightcoral'])
plt.title('Conversion Rates by Marketing Channel Type')
plt.xlabel('Marketing Channel Type')
plt.ylabel('Conversion Rate')
plt.ylim(0, max(conversion_rates['conversion_rate']) * 1.2) # Set y-limit slightly above max for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("Bar plot showing conversion rates by marketing channel type has been generated.")
````

## Conduct Statistical Analysis and Visualization

### Subtask:
Execute a two-sample proportion z-test to compare conversion rates between 'paid' and 'other' marketing channels, interpret the results, and ensure visualizations support the findings.


**Reasoning**:
To perform the two-sample proportion z-test, I will first extract the necessary counts and totals from the `conversion_rates` DataFrame, then import the `proportions_ztest` function, and execute the test with the specified alternative hypothesis.



````python
from statsmodels.stats.proportion import proportions_ztest

# Extract data for 'paid' and 'other' channels from conversion_rates DataFrame
paid_sessions = conversion_rates[conversion_rates['marketing_channel_type'] == 'paid']['total_sessions'].iloc[0]
paid_conversions = conversion_rates[conversion_rates['marketing_channel_type'] == 'paid']['total_conversions'].iloc[0]

other_sessions = conversion_rates[conversion_rates['marketing_channel_type'] == 'other']['total_sessions'].iloc[0]
other_conversions = conversion_rates[conversion_rates['marketing_channel_type'] == 'other']['total_conversions'].iloc[0]

# Combine counts and observations for the z-test
counts = [paid_conversions, other_conversions]
nobs = [paid_sessions, other_sessions]

# Define significance level
alpha = 0.05

# Perform two-sample proportion z-test (alternative='larger' because H1 is paid > other)
z_statistic, p_value = proportions_ztest(counts, nobs, alternative='larger')

print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Compare p-value to significance level and state conclusion
if p_value < alpha:
    print(f"\nSince the p-value ({p_value:.4f}) is less than the significance level (alpha = {alpha}), we reject the Null Hypothesis.")
    print("Conclusion: There is a statistically significant evidence that paid marketing channels lead to a higher conversion rate compared to other marketing channels.")
else:
    print(f"\nSince the p-value ({p_value:.4f}) is greater than or equal to the significance level (alpha = {alpha}), we fail to reject the Null Hypothesis.")
    print("Conclusion: There is no statistically significant evidence that paid marketing channels lead to a higher conversion rate compared to other marketing channels.")
````

## Analyze A/B Test: Desktop vs. Mobile Conversion Rate

### Subtask:
Formulate a business hypothesis and corresponding statistical hypotheses (H₀, H₁) to compare conversion rates between desktop and mobile users. Prepare the data by extracting relevant information from the `df_merged` DataFrame.


### Business Hypothesis

**Hypothesis:** Desktop users will exhibit a higher conversion rate to orders compared to mobile users.

**Rationale:** This hypothesis aligns with the task's focus on analyzing A/B tests for different user segments (desktop vs. mobile) and is clear, testable, and specific using the available `device_type` and `converted` columns in the `df_merged` DataFrame.

### Statistical Hypotheses

1.  **Null Hypothesis (H₀):** There is no significant difference in conversion rates between desktop users and mobile users. (i.e., The conversion rate of desktop users is less than or equal to the conversion rate of mobile users).
    
    *Mathematically:* $\mu_{\text{desktop}} \le \mu_{\text{mobile}}$

2.  **Alternative Hypothesis (H₁):** Desktop users lead to a significantly higher conversion rate compared to mobile users.
    
    *Mathematically:* $\mu_{\text{desktop}} > \mu_{\text{mobile}}$

### Significance Level:

*   **Alpha (α):** 0.05 (as recommended in the brief and previous analyses).

This means we are willing to accept a 5% chance of incorrectly rejecting the Null Hypothesis (Type I error).

**Reasoning**:
Now that the hypotheses are defined, I will filter the `df_merged` DataFrame to separate sessions by 'desktop' and 'mobile' device types and calculate the total sessions and conversions for each, storing these values for further analysis as per instructions 3, 4, and 5.



````python
desktop_sessions = df_merged[df_merged['device_type'] == 'desktop']
mobile_sessions = df_merged[df_merged['device_type'] == 'mobile']

desktop_total_sessions = len(desktop_sessions)
desktop_total_conversions = desktop_sessions['converted'].sum()

mobile_total_sessions = len(mobile_sessions)
mobile_total_conversions = mobile_sessions['converted'].sum()

print(f"Desktop Sessions: {desktop_total_sessions}, Desktop Conversions: {desktop_total_conversions}")
print(f"Mobile Sessions: {mobile_total_sessions}, Mobile Conversions: {mobile_total_conversions}")
````

**Reasoning**:
Now that the total sessions and conversions for desktop and mobile users have been calculated, the next step is to perform a two-sample proportion z-test to compare their conversion rates, interpreting the p-value against the defined significance level (α = 0.05).



````python
from statsmodels.stats.proportion import proportions_ztest

# Combine counts and observations for the z-test
counts = [desktop_total_conversions, mobile_total_conversions]
nobs = [desktop_total_sessions, mobile_total_sessions]

# Define significance level
alpha = 0.05

# Perform two-sample proportion z-test (alternative='larger' because H1 is desktop > mobile)
z_statistic, p_value = proportions_ztest(counts, nobs, alternative='larger')

print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Compare p-value to significance level and state conclusion
if p_value < alpha:
    print(f"\nSince the p-value ({p_value:.4f}) is less than the significance level (alpha = {alpha}), we reject the Null Hypothesis.")
    print("Conclusion: There is a statistically significant evidence that desktop users lead to a higher conversion rate compared to mobile users.")
else:
    print(f"\nSince the p-value ({p_value:.4f}) is greater than or equal to the significance level (alpha = {alpha}), we fail to reject the Null Hypothesis.")
    print("Conclusion: There is no statistically significant evidence that desktop users lead to a higher conversion rate compared to mobile users.")
````

**Reasoning**:
To prepare for visualization and summarize the comparison, I will calculate the conversion rates for both desktop and mobile users and store them in a DataFrame, which will make it easy to create a bar plot to visually represent the findings.



````python
desktop_conversion_rate = desktop_total_conversions / desktop_total_sessions
mobile_conversion_rate = mobile_total_conversions / mobile_total_sessions

conversion_rates_by_device = pd.DataFrame({
    'device_type': ['desktop', 'mobile'],
    'total_sessions': [desktop_total_sessions, mobile_total_sessions],
    'total_conversions': [desktop_total_conversions, mobile_total_conversions],
    'conversion_rate': [desktop_conversion_rate, mobile_conversion_rate]
})

print("Conversion rates by device type:")
print(conversion_rates_by_device)
````

**Reasoning**:
To visually represent the conversion rates by device type, I will create a bar plot using the `conversion_rates_by_device` DataFrame. This will clearly illustrate the difference in conversion rates between desktop and mobile users, supporting the statistical analysis.



````python
import matplotlib.pyplot as plt

# Visualize the conversion rates for desktop vs. mobile users
plt.figure(figsize=(8, 6))
plt.bar(conversion_rates_by_device['device_type'], conversion_rates_by_device['conversion_rate'], color=['darkblue', 'orange'])
plt.title('Conversion Rates by Device Type')
plt.xlabel('Device Type')
plt.ylabel('Conversion Rate')
plt.ylim(0, max(conversion_rates_by_device['conversion_rate']) * 1.2) # Set y-limit slightly above max for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("Bar plot showing conversion rates by device type has been generated.")
````

## Analyze A/B Test: New vs. Returning Customer Average Order Value

### Subtask:
Formulate a business hypothesis and statistical hypotheses (H₀, H₁) to compare the average order value (AOV) between new and returning customers. Prepare the data by filtering `df_merged` to identify new and returning sessions and link them to order values.


### Business Hypothesis

**Hypothesis:** Returning customers will exhibit a higher Average Order Value (AOV) compared to new customers.

**Rationale:** This hypothesis aligns with common e-commerce patterns where established customers often have higher trust and loyalty, leading to larger purchases. The `is_repeat_session` column in `df_merged` can identify returning customers, and `price_usd` from `orders_df` will be used to calculate AOV.

### Statistical Hypotheses

1.  **Null Hypothesis (H₀):** There is no significant difference in Average Order Value (AOV) between new and returning customers. (i.e., The AOV of returning customers is less than or equal to the AOV of new customers).
    
    *Mathematically:* $\mu_{\text{returning}} \le \mu_{\text{new}}$

2.  **Alternative Hypothesis (H₁):** Returning customers lead to a significantly higher Average Order Value (AOV) compared to new customers.
    
    *Mathematically:* $\mu_{\text{returning}} > \mu_{\text{new}}$

### Significance Level:

*   **Alpha (α):** 0.05 (as recommended in the brief and previous analyses).

This means we are willing to accept a 5% chance of incorrectly rejecting the Null Hypothesis (Type I error).

**Reasoning**:
Now that the hypotheses are defined, I will merge the `df_merged` DataFrame with the `orders_df` DataFrame using 'order_id' to bring in the `price_usd` for each order, as specified in instruction 3.



````python
df_merged_with_orders = pd.merge(
    df_merged,
    orders_df[['order_id', 'price_usd']],
    on='order_id',
    how='left'
)

print("Merged df_merged with orders_df to include price_usd.")
print("Head of df_merged_with_orders:")
print(df_merged_with_orders.head())
````

**Reasoning**:
I will filter the `df_merged_with_orders` DataFrame into two separate DataFrames, `new_customer_sessions` and `returning_customer_sessions`, based on the `is_repeat_session` column, as instructed by the subtask.



````python
new_customer_sessions = df_merged_with_orders[df_merged_with_orders['is_repeat_session'] == 0]
returning_customer_sessions = df_merged_with_orders[df_merged_with_orders['is_repeat_session'] == 1]

print(f"Number of new customer sessions: {len(new_customer_sessions)}")
print(f"Number of returning customer sessions: {len(returning_customer_sessions)}")

print("\nHead of new_customer_sessions:")
print(new_customer_sessions.head())
print("\nHead of returning_customer_sessions:")
print(returning_customer_sessions.head())
````

**Reasoning**:
Now that the sessions are separated by new and returning customers, I will calculate the Average Order Value (AOV) for each group by filtering for converted sessions and then taking the mean of the 'price_usd' column, as specified in instruction 5.



````python
new_customer_aov = new_customer_sessions[new_customer_sessions['converted'] == 1]['price_usd'].mean()
returning_customer_aov = returning_customer_sessions[returning_customer_sessions['converted'] == 1]['price_usd'].mean()

print(f"Average Order Value for New Customers: ${new_customer_aov:.2f}")
print(f"Average Order Value for Returning Customers: ${returning_customer_aov:.2f}")
````

**Reasoning**:
Now that the Average Order Value (AOV) for new and returning customers has been calculated, I will perform a two-sample t-test to compare these means, as per the established statistical hypotheses and the need to determine statistical significance.



````python
from scipy import stats

# Extract price_usd for converted new and returning customer sessions
new_customer_orders = new_customer_sessions[new_customer_sessions['converted'] == 1]['price_usd'].dropna()
returning_customer_orders = returning_customer_sessions[returning_customer_sessions['converted'] == 1]['price_usd'].dropna()

# Define significance level
alpha = 0.05

# Perform independent two-sample t-test (alternative='greater' because H1 is returning > new)
t_statistic, p_value = stats.ttest_ind(returning_customer_orders, new_customer_orders, equal_var=False, alternative='greater')

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Compare p-value to significance level and state conclusion
if p_value < alpha:
    print(f"\nSince the p-value ({p_value:.4f}) is less than the significance level (alpha = {alpha}), we reject the Null Hypothesis.")
    print("Conclusion: There is statistically significant evidence that returning customers have a higher Average Order Value compared to new customers.")
else:
    print(f"\nSince the p-value ({p_value:.4f}) is greater than or equal to the significance level (alpha = {alpha}), we fail to reject the Null Hypothesis.")
    print("Conclusion: There is no statistically significant evidence that returning customers have a higher Average Order Value compared to new customers.")
````

````python
import matplotlib.pyplot as plt

# Create a DataFrame for visualization of AOV
aov_data = pd.DataFrame({
    'customer_type': ['New Customers', 'Returning Customers'],
    'aov': [new_customer_aov, returning_customer_aov]
})

# Visualize the AOV for new vs. returning customers
plt.figure(figsize=(8, 6))
plt.bar(aov_data['customer_type'], aov_data['aov'], color=['lightblue', 'darkgreen'])
plt.title('Average Order Value (AOV) by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Average Order Value (USD)')
plt.ylim(0, max(aov_data['aov']) * 1.2) # Set y-limit slightly above max for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("Bar plot showing Average Order Value by customer type has been generated.")
````

### 2. Analysis 1: Paid vs. Other Marketing Channels Conversion Rate

**Business Hypothesis:** Sessions originating from 'gsearch' (Google Search) will exhibit a higher conversion rate to orders compared to sessions originating from 'bsearch' (Bing Search), or more broadly, 'paid' channels will have a higher conversion rate than 'other' channels.

**Statistical Hypotheses:**
*   **Null Hypothesis (H₀):** There is no significant difference in conversion rates between paid marketing channels and other marketing channels ($\mu_{\text{paid}} \le \mu_{\text{other}}$).
*   **Alternative Hypothesis (H₁):** Paid marketing channels lead to a significantly higher conversion rate compared to other marketing channels ($\mu_{\text{paid}} > \mu_{\text{other}}$).
*   **Significance Level (α):** 0.05.

**Analysis Steps:**
1.  The `website_sessions` and `orders` DataFrames were loaded, and their `created_at` columns were converted to datetime objects.
2.  The `website_sessions_df` was merged with `orders_df` (specifically `order_id`) using a left merge on `website_session_id` to link sessions to orders.

6.  A bar plot visualized the distribution of marketing channel types.
7.  A bar plot displayed the conversion rates for 'paid' vs. 'other' channels.
8.  A two-sample proportion z-test was conducted to statistically compare the conversion rates, using `alternative='larger'` based on H₁.

**Key Findings:**

    *   'Paid' channels: 414,060 sessions, 28,371 conversions, **Conversion Rate = 6.85%**.

*   **Statistical Significance:** The two-sample proportion z-test yielded a z-statistic of 1.3408 and a **p-value of 0.0900**. Since this p-value (0.0900) was greater than the significance level (α = 0.05), we failed to reject the Null Hypothesis. This indicates that there was no statistically significant evidence to conclude that paid marketing channels lead to a higher conversion rate compared to other marketing channels based on the provided data.
